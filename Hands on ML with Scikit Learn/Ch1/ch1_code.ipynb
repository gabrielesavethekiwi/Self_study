There are two approaches to learning, which can be
- Model based
- Instance Based

In *model based learning* you develop an algorithm that learns the parameters of a model, such as the slope and intercept for a linear regression. Once the model is calibrated, we do not even need to retain past data.
By contrast, in *instance based* learning, we define a notion of distance between data points. Therefore, if we are given a new data point and we want to predict the value of a target quantity, we will at first compute its "distance" from the other data points. k nearest neighbours, with k = 1, is an example of instance based learning where the value of a target quantity is inferred by looking at the closest data point in our training set (or closest k points, then we average over them).
